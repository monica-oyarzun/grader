```python
import pandas as pd
```

```python
df = pd.read_csv('data/grades.csv')
df.head(2)
```

```python
df1 = df.iloc[:,[0,1,3,5]]
df1.head(2)

# this gives us columns at listed indices... easier than typing out long names and using .loc
```

```python
# Let's assume grade3 is a final test and it counts double

df1.apply(lambda z: print(type(z)))

# What does this give you? A Series for each column....
```

```python
df1.apply(lambda z: print(len(z)))
```

```python
# We want to sum across rows, and make final score equal to twice the score:

df1.apply(lambda row: row[1] + row[2] + (2*row[3]), axis=1)
```

```python
# Now to get the average, divide by 400 (total points):
# 'row' here is the parameter function
# ** axis=1 is what tells us to add across rows instead of columns
# by default... axis will set to 0 and that sums (or does whatever) down columns, axis = 1 goes across rows

df1['final'] = df1.apply(lambda row: (row[1] + row[2] + (2*row[3]))/400, axis=1)
```

```python
df1.head(2)
```

```python
# want to find top 1% of class:

num_students = int(len(df) * .01)
num_students
```

```python
# this will sort students by score:

indices = df1.final.argsort()
indices.head()
```

```python
# see the first person (1424) that is in the list... they are the lowest scoring person:

df1.final[1424]
```

```python
df1.iloc[indices]

# this is now sorted from lowest to highest
```

```python
df1.iloc[indices[-num_students:]]

# this gets just the last 23 students, who scored the highest
```

```python
filtered_students = df1.iloc[indices[-num_students:]][::-1]
filtered_students

# this reverses to get students in order from highest score to lowest
```
